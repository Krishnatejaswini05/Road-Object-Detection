```python
from imageai.Detection import VideoObjectDetection
#from imageai.Detection import ObjectDetection
import os
from matplotlib import pyplot as plt
import cv2
from timeit import default_timer as time
from keras.models import load_model
from keras import backend as K
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
import requests
import pandas as pd
import re
import time
import json
from imageai.Detection import ObjectDetection

execution_path = os.getcwd()
detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
#detector.setModelTypeAsTinyYOLOv3()
detector.setModelPath(os.path.join(execution_path,"yolo.h5"))
#detector.setModelPath(os.path.join(execution_path,"helmet-nonhelmet_cnn.h5"))
#detector.setModelPath(os.path.join(execution_path,"yolov3.pt"))
detector.loadModel()

custom =  detector.CustomObjects(person = True)
#custom =  detector.CustomObjects("Motorcycle")


#custom =  detector.CustomObjects(person = True)
plt.show()

#detections = detector.detectCustomObjectsFromImage(

detections = detector.detectObjectsFromImage(
        custom_objects= custom,
        input_image = os.path.join(execution_path,"triple.jpg"),
        output_image_path = os.path.join(execution_path,"traffic-detected3.jpeg"),
        minimum_percentage_probability = 30

)
count =0
imjh = load_img("triple.jpg")
plt.figure()
plt.imshow(imjh)
plt.show()
for eachObject in detections:
    print(eachObject["name"]," : ",eachObject["percentage_probability"], " : ",eachObject["box_points"] )
    print("------------------------------------")
    count+=1
print(count)
imkh = load_img("traffic-detected3.jpeg")
plt.figure()
plt.imshow(imkh)
plt.show()
```
